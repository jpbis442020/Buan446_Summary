# Q1
Exam Question 1: Write the code to recreate the table below. You can do it in a few different ways but the easiest is to us a for loop and the f-string capabilities. The numbers should be right aligned in each column.



```python
# Question 1 Code (6 points total)
# 3 Points for table containing the appropriate data
# 2 Points for table similar to the picture
# 1 Points for right aligning the number in each colum
def squaresandcubes():
    print("Number\tSquare\tCube")
    print("---------------------")
    for number in range(11):
        square = number ** 2
        cube = number ** 3
        print(f"{number:>6}\t{square:>6}\t{cube:>4}")
squaresandcubes()
```

# Q4
Create a list called numbers with the following [2, 14, -3, 9, 0, 8, -2]
Iterate through the list with a for loop and evaluate each item an print if the number is positive, negative or zero. It will look similar to this image

```python
# Code for Problem 4 (6 points total)
# Create the list numbers 1 point
# Create a for loop to iterate through the list numbers 2 points
# Within the for loop print positive, negative or zero 3 points
# Prompt the user for an integer

l=[2, 14, -3, 9, 0, 8, -2]
# Check if the number is odd or even

for i in l:
    if i>0:
        print(i, "is a positive number.")
        
    elif i<0:
        print(i, "is a negative number.")
        
    else:
        print(i, "is zero.")
```

Take the code in Q4 and add a counter to automatically display the total number of positive, negative and Zeros. Result should look similar to below
